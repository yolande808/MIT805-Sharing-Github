```python
#import necessary packages
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
jan = pd.read_csv("yellow_tripdata_2019-01.csv", usecols=range(1,17))
```

```python
jan.head()
```

```python
jan = jan.to_csv('jan.csv', float_format='%.3f',index=False)
```

```python
feb = pd.read_csv("yellow_tripdata_2019-02.csv", usecols=range(1,17))
```

```python
mar = pd.read_csv("yellow_tripdata_2019-03.csv", usecols=range(1,17))
```

```python
Final = pd.concat([jan,feb,mar], ignore_index=True)
```

```python
Final.head()
```

```python
pick_up_time = Final["tpep_pickup_datetime"]
pick_up_time.head()
```

```python
from datetime import datetime
DateTime = pd.to_datetime(pick_up_time, errors='coerce')
```

```python
#create a new variable pick up date in our dataset
Final["pick_up_date_time"] = DateTime
```

```python
Final.head()
```

```python
Final = Final[(Final['pick_up_date_time'].dt.year <= 2020)]
```

```python
#drop the months 4..12 
Final = Final[(Final['pick_up_date_time'].dt.month <= 3)]              
```

```python
Final["pick_up_month"] = DateTime.dt.month
```

```python
Final["pick_up_day"] = DateTime.dt.day
```

```python
Final["pick_up_hour"] = DateTime.dt.hour
```

```python
Final.head()
```

```python
drop_off_time = Final["tpep_dropoff_datetime"]
```

```python
DateTime_drop = pd.to_datetime(drop_off_time, errors='coerce')
```

```python
#create a new variable drop off date in our dataset
Final["drop_off_date_time"] = DateTime_drop
```

```python
Final.head()
```

```python
#trip duration in minutes
duration = (Final.drop_off_date_time - Final.pick_up_date_time).dt.seconds
```

```python
trip_duration_min = duration/60
#print(duration)
```

```python
#create a new variable duration in our dataset
Final["trip duration"] = trip_duration_min
```

```python
final_data = Final.to_csv('final.csv', float_format='%.2f',index=False, header = False)
```
